```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

# Load the Iris dataset
data = load_iris()
X, y = data['data'], data['target']

# Create an imbalanced dataset by duplicating samples from the minority class
np.random.seed(42)
minority_class = 2
minority_samples = X[y == minority_class]
minority_labels = y[y == minority_class]
num_minority_samples = 10

X_imbalanced = np.concatenate([X, np.tile(minority_samples, (num_minority_samples, 1))], axis=0)
y_imbalanced = np.concatenate([y, np.tile(minority_labels, num_minority_samples)], axis=0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imbalanced, y_imbalanced, test_size=0.2, random_state=42)

# Display class distribution in the training set before resampling
unique_classes, class_counts = np.unique(y_train, return_counts=True)
print("Class distribution in the training set before resampling:")
print(pd.DataFrame({"Class": unique_classes, "Count": class_counts}))
```

```python
X_resampled, y_resampled = resample(X_train[y_train != minority_class], y_train[y_train != minority_class],
                                    n_samples=class_counts.max(), random_state=42)
X_train_resampled = np.vstack([X_train[y_train == minority_class], X_resampled])
y_train_resampled = np.concatenate([y_train[y_train == minority_class], y_resampled])
```

```python
unique_classes_resampled, class_counts_resampled = np.unique(y_train_resampled, return_counts=True)
print("\nClass distribution in the training set after resampling:")
print(pd.DataFrame({"Class": unique_classes_resampled, "Count": class_counts_resampled}))
```
